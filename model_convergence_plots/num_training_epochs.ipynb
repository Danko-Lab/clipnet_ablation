```python
import pandas as pd
import seaborn as sns
from scipy.stats import mannwhitneyu, ttest_ind
import matplotlib.pyplot as plt
import os
os.makedirs("img", exist_ok=True)
```

```python
data = pd.DataFrame(
    {
        "CLIPNET": [69, 54, 63, 63, 42, 39, 29, 29, 43],
        "CLIPNET-ref": [66, 55, 38, 16, 63, 50, 47, 59, 15],
    }
)
```

```python
# violin plot, add dots with jitter
fig = sns.violinplot(data=data, alpha=0.1)

# add dots with jitter
sns.stripplot(data=data, jitter=True, size=6)

plt.ylabel("Number of training epochs before early stopping")
plt.savefig("img/clipnet_num_epochs.pdf")
```

```python
mannwhitneyu(data.ref_num_epochs, data.pers_num_epochs, alternative="less")
```

```python
ttest_ind(data.ref_num_epochs, data.pers_num_epochs, alternative="less")
```
