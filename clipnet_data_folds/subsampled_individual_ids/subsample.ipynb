```python
import pandas as pd
import random
import numpy as np
```

```python
df = pd.read_csv("procap_library_prefixes.txt", header=None, names=["expt_id"])
df["individual_id"] = df["expt_id"].str.split("_").str[0]
```

```python
subsample_n = 50
n_runs = 5

for run in range(n_runs):
    subsample_expt_ids = []
    subsample_individual_ids = []
    # Sample expt ids so that no individual_id is sampled more than once
    while len(subsample_expt_ids) < subsample_n:
        sample = df.sample(1)
        if sample["individual_id"].values[0] not in subsample_individual_ids:
            subsample_expt_ids.append(sample["expt_id"].values[0])
            subsample_individual_ids.append(sample["individual_id"].values[0])

    subsample_expt_ids = sorted(subsample_expt_ids)
    subsample_individual_ids = sorted(subsample_individual_ids)

    with open(f"subsample_prefixes_n{subsample_n}_run{run}.txt", "w") as f:
        f.write("\n".join(subsample_expt_ids))
```
